```python
def parse_soup(soup):
    soup_str = str(soup)
    soup_words = soup_str.split("\"")
    values = []
    for idx, word in enumerate(soup_words):
        if word == 'studentsPerTeacher':
            value = int(soup_words[idx+1].strip(':').strip(','))
            values.append(value)
    return np.mean(values)
```

```python
def get_teacher_ratio(state, town):
    state = state.replace(' ', '-')
    town = town.replace(' ', '-')
    try:
        url = f'https://www.greatschools.org/{state}/{town}/schools/?view=table'
        res = requests.get(url)
        soup = BeautifulSoup(res.content, 'lxml')
        return parse_soup(soup)
    except:
        return np.nan
```

```python
# Loop through each town in each state to scrape crime data
for state in states:
    if state not in done_states:
        towns = list(set(inc.loc[inc['state'] == state, 'town']))
        for town in towns:
            inc.loc[(inc['state'] == state) & (inc['town'] == town), 'students_per_teacher'] = get_teacher_ratio(state, town)
            pd.to_pickle(inc, '/Users/elijahcurme/Desktop/GA/capstone/pkl/education.pkl')
        done_states.append(state)
    print(f"{state} finished")
```
